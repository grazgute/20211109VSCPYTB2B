# Seaborn

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.histplot(data = tips, x="size")
```

```python
sns.histplot(data = tips, x="size", hue = "smoker")
```

```python
sns.jointplot(data = tips, x="size", y="total_bill", hue = "sex")
```

```python
sns.lmplot(data = tips, x="size", y="total_bill", hue = "sex", col="day")
```

```python
sns.countplot(data=tips, x='day', hue='sex')
```
